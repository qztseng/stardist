## plot Huber loss

```python
def huber_loss(error, delta=0.05):
    abs_error = np.abs(error)
    quadratic = np.minimum.outer(abs_error, delta)
    linear = abs_error - quadratic
    return 0.5 * np.square(quadratic) + delta * linear
```

```python
huber_loss(0.9907, 0.05)*np.exp(0.4907*3)
```

```python
x = np.linspace(0,1,100)
dd = np.linspace(0.1, 1,5)
# dd = [ 0.1, 0.2, 0.5, 0.8]
dfs = []
for d in dd:
#     print(x, d)
    y = huber_loss(x, d)
    df = pd.DataFrame(data={'x':x, 'y':y, 'delta':d})
    dfs.append(df)
# dfs = pd.concat(dfs)
# dfs.delta = dfs.delta.astype('category')
# dfs
```

```python

# style
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')
plt.figure(figsize=(10,10))
# multiple line plot
num=0
for df in dfs:
    num+=1
    plt.plot(df['x'], df['y'], marker='', color=palette(num), linewidth=2, alpha=1, label=df['delta'].unique().item())

ycub = np.power(df.x,1.5)
plt.plot(df.x, ycub, marker='', color=palette(num+1), linewidth=2, alpha=1, label='cubic')
ysqr = np.power(df.x,2)
plt.plot(df.x, ysqr, marker='', color=palette(num+2), linewidth=2, alpha=1, label='sqr')
# Add legend
plt.legend(loc=2, ncol=2)

# Add titles
plt.title("Huber loss for x<1", loc='left', fontsize=12, fontweight=4)
plt.xlabel("error")
plt.ylabel("huber loss")
```

## plot negative exponential function with 0<=x<=1

```python
x = np.linspace(0,1,100)
delta = np.linspace(0.1,1,10)
dfs = []
for d in delta:
    y = np.exp(-x/d)
    df = pd.DataFrame(data={'x':x, 'y':y, 'delta':d})
    dfs.append(df)
```

```python
# plt.style.available
```

```python
plt.style.use('ggplot') 
# create a color palette
palette = plt.get_cmap('Set1')
plt.figure(figsize=(10,10))
# multiple line plot
num=0
for df in dfs:
    num+=1
    plt.plot(df['x'], df['y'], marker='', color=palette(num), linewidth=2, alpha=1, label=df['delta'].unique().item())

# Add legend
plt.legend(loc='upper right', ncol=2)

# Add titles
plt.title("exp(-(y-Yhat)/delta)", loc='left', fontsize=16, fontweight=2)
plt.xlabel("y - Yhat")
plt.ylabel("transformed")  
```

```python
x = np.linspace(np.pi,4*np.pi,200)

palette = plt.get_cmap('tab10')
plt.figure(figsize=(10,10))
for i,m in enumerate(np.arange(0,1,0.1)):
    y = 0.5*(special.ellipj(x, m)[0]+1)
    plt.plot(x, y, color=palette(i), label=f'{m:.1f}')
plt.legend( bbox_to_anchor=(1, 1))
```
